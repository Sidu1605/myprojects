## Outliers

```python
#1) checking outliers
def check_outliers():
    global df_train,df_test
    continuous_features=[i for i in df_train.columns if (df_train[i].dtype!='O' and df_train[i].value_counts().count()>15) and (i != 'Property_Sale_Price')]    
    data_train=df_train.copy()
    data_test=df_test.copy()
    for feature in continuous_features:
        #check outliers train
        sns.boxplot(data_train[feature])
        plt.title(feature)
        plt.show()
        #check outliers test
        sns.boxplot(data_test[feature])
        plt.title(feature+' test',fontsize=20)
        plt.show()

#2) treating outliers with winsorization
def winsorization(lower_limit,upper_limit):
    global df_train,df_test,out_train,out_test
    out_train=df_train.copy()
    out_test=df_test.copy()
    continuous_features=[i for i in df_train.columns if (df_train[i].dtype!='O' and df_train[i].value_counts().count()>15) and (i != 'Property_Sale_Price')]
    for feature in continuous_features:
        out_train[feature]=winsorize(out_train[feature], limits=[lower_limit,upper_limit])
        sns.boxplot(out_train[feature])
        plt.title(feature+' with train',fontsize=20)
        plt.show()
        out_test[feature]=winsorize(out_test[feature], limits=[lower_limit,upper_limit])
        sns.boxplot(out_test[feature])
        plt.title(feature+' with test',fontsize=20)
        plt.show()
```

## Missing Values

```python
#1) display missing value tables
#displaying missing values features details
def missing_table():
    global out_train,out_test,target,miss_train,miss_test,missing_train,missing_test
    miss_train=out_train.copy()
    miss_test=out_test.copy()
    missing_train,missing_test=pd.DataFrame(),pd.DataFrame()
    #total missing values count
    missing_train['Total'] = miss_train.isnull().sum().sort_values(ascending = False)
    missing_test['Total'] = miss_test.isnull().sum().sort_values(ascending = False)
    #
    missing_train['Percentage of Missing Values'] = (miss_train.isnull().mean()*100).sort_values(ascending = False)   
    missing_test['Percentage of Missing Values'] = (miss_test.isnull().mean()*100).sort_values(ascending = False)   
    
    missing_train['Type'] = miss_train[missing_train.index].dtypes
    missing_test['Type'] = miss_test[missing_test.index].dtypes
    
    missing_train['not_nul_mean-null_mean']=np.nan
    missing_test['not_nul_mean-null_mean']=np.nan
    
    missing_train=missing_train[missing_train['Percentage of Missing Values']>0]
    missing_test=missing_test[missing_test['Percentage of Missing Values']>0]
    
    missing_train['num_of_unique_class'] = miss_train[missing_train.index].nunique()
    missing_test['num_of_unique_class'] = miss_test[missing_test.index].nunique()
    
    x_features=[i for i in miss_train.columns if i !=target]
    
    data_train1=np.where(miss_train[x_features].isnull(),1,0)
    data_train1=pd.DataFrame(data_train1,columns=x_features)
    df_train_y1=miss_train[target].reset_index(level=None, drop=False, name=None, inplace=False)
    data_train11=data_train1.join(df_train_y1, on=None, how='inner')
    
    for i in range(0,missing_train.count()[0]):
        col=missing_train['not_nul_mean-null_mean'].index[i]
        col_group=data_train11.groupby(col)[target].median()
        comp_value=col_group[0]-col_group[1]
        missing_train.iloc[i,3]=comp_value
    
    data_test1=np.where(miss_test[x_features].isnull(),1,0)
    data_test1=pd.DataFrame(data_test1,columns=x_features)
    df_test_y1=miss_test[target].reset_index(level=None, drop=False, name=None, inplace=False)
    data_test11=data_test1.join(df_test_y1, on=None, how='inner')
    
    for i in range(0,missing_test.count()[0]):
        col=missing_test['not_nul_mean-null_mean'].index[i]
        col_group=data_test11.groupby(col)[target].median()
        comp_value=col_group[0]-col_group[1]
        missing_test.iloc[i,3]=comp_value

#2) check difference between null and not null value in independent variable Using dependent variable mean 
#creating function for how much null value features impacting on target feature
def missing_plot():
    global missing_trian,miss_train,miss_test,target
    df_missing=missing_train.index
    data_train=miss_train.copy()
    data_test=miss_test.copy()
    for i in df_missing:
        data_train[i]=np.where(data_train[i].isnull(),1,0)
        data_train.groupby(i)[target].mean().plot.bar()
        plt.title(i+' with '+target+' in training dataset',fontsize=15)
        plt.show()
    print('\n')
    for i in df_missing:
        data_test[i]=np.where(data_test[i].isnull(),1,0)
        data_test.groupby(i)[target].mean().plot.bar()
        plt.title(i+' with '+target+' in testing dataset',fontsize=15)
        plt.show()

#3) ploting heatmap for missing values
def missing_heatmap():
    global miss_train,miss_test
    #training dataset
    sns.heatmap(miss_train.isnull(), cbar=False)
    plt.title('Missing values in training dataset',fontsize=20)
    plt.show()
    #testing dataset
    sns.heatmap(miss_train.isnull(), cbar=False)
    plt.title('Missing values in testing dataset',fontsize=20)
    plt.show()

#4) handling missing values

#creating function for to handle missing values categorical & numarical datatype featues
def handling_missing():
    global miss_train,miss_test,missing_train,df_train9,df_test9
    df_train9=miss_train.copy()
    df_test9=miss_test.copy()
    mis=missing_train.copy()
    for i in range(0,mis.count()[0]):
        x1=mis.iloc[i,1]    #Percentage of Missing Values4
        x2=mis.iloc[i,2]    #Type
        x3=mis.iloc[i,3]    #not_nul_mean-null_mean
        x4=mis.iloc[i,4]    #num_of_unique_class
        col=mis.index[i]     #column name
        if x2!='object' and df_train9[col].isnull().sum()>0:
            if x4>=15:
                if x1>=40:
                    if x3>0:
                        df_train9.drop(col,axis=1,inplace=True)
                        df_test9.drop(col,axis=1,inplace=True)
                        print(col,'feature is dropped')
                    else:
                        df_train9[col].fillna(df_train9[col].mean(), inplace=True)
                        df_test9[col].fillna(df_test9[col].mean(), inplace=True)
                        print(col,'feature is filled with mean')
                elif x1>=1:
                    df_train9[col].fillna(df_train9[col].mean(), inplace=True)
                    df_test9[col].fillna(df_test9[col].mean(), inplace=True)
                    print(col,'feature is filled with mean')
                else:
                    if x3>0:
                        df_train9=df_train9[~(df_train9[col].isnull())]
                        df_test9=df_test9[~(df_test9[col].isnull())]
                        print(col,'feature sclising not null records.')
                    else:
                        df_train9[col].fillna(df_train9[col].mean(), inplace=True)
                        df_test9[col].fillna(df_test9[col].mean(), inplace=True)
                        print(col,'feature is filled with mean.')
            else:
                if x1>=40:
                    if x3>0:
                        df_train9.drop(col,axis=1,inplace=True)                     
                        df_test9.drop(col,axis=1,inplace=True)
                        print(col,'feature is dropped')
                    else:
                        df_train9[col].fillna(df_train9[col].median(), inplace=True)
                        df_test9[col].fillna(df_test9[col].median(), inplace=True)
                        print(col,'feature is filled with median.')
                elif x1>=1:
                    df_train9[col].fillna(df_train9[col].median(), inplace=True)
                    df_test9[col].fillna(df_test9[col].median(), inplace=True)
                    print(col,'feature is filled with median.')
                else:
                    if x3>0:
                        df_train9[col].fillna(df_train9[col].median(), inplace=True)
                        df_test9[col].fillna(df_test9[col].median(), inplace=True)
                        print(col,'feature is filled with median.')
                    else:
                        df_train9=df_train9[~(df_train9[col].isnull())]
                        df_test9=df_test9[~(df_test9[col].isnull())]
                        print(col,'feature null value rows are dropped')
        else:
            if x1>=40 and df_train9[col].isnull().sum()>0:                  #drop columns
                if x3>0:
                    df_train9.drop(col,axis=1,inplace=True)
                    df_test9.drop(col,axis=1,inplace=True)
                    print(col,'feature is dropped')
                else:
                    df_train9[col].fillna(df_train9[col].mode()[0], inplace=True)   #replace Unkown value
                    df_test9[col].fillna(df_test9[col].mode()[0], inplace=True)
                    print(col,'feature is filled with mode.')
            elif x1>=1 and df_train9[col].isnull().sum()>0:
                if 'Other' in df_train9[col].unique():
                    df_train9[col].fillna('Other', inplace=True)      #replace Other value
                elif 'NA' in df_train9[col].unique():
                    df_train9[col].fillna('NA', inplace=True)      #replace NA value
                else:
                    df_train9[col].fillna('Unknown', inplace=True)      #replace Unkown value
                    df_test9[col].fillna('Unknown', inplace=True)
                    print(col,'feature is filled with mode.')
            else:
                if x3>0 and df_train9[col].isnull().sum()>0:
                    df_train9[col].fillna(df_train9[col].mode()[0], inplace=True)  #replace mode value
                    df_test9[col].fillna(df_test9[col].mode()[0], inplace=True)
                    print(col,'feature null values filled with mode.')
                else:
                    if df_train9[col].isnull().sum()>0:
                        df_train9=df_train9[~(df_train9[col].isnull())]
                        df_test9=df_test9[~(df_test9[col].isnull())]
                        print(col,'feature null values slicied not null values.')
                    else:
                        print('check missing values in ',col,'feature.')
```

## Multi-Coolinearity

```python
## multi-colinearity
#function for co-relation heatmap
def co_check(threshold):
    global df_train9,df_test9,target,co_train,co_test
    co_train,co_test=df_train9.copy(),df_test9.copy()
    df_train_num=co_train[[i for i in co_train.columns if co_train[i].dtype!='O']]
    corr=df_train_num.drop(target,axis=1).corr()
    sns.heatmap(corr[(corr>=threshold)|(corr<=-(threshold))],cmap='viridis',vmax=1.0,linewidths=.1,annot=True,annot_kws={'size':10},square=True);
    plt.title('co-relation plot using training dataset')
    plt.show()
    df_test_num=co_test[[i for i in co_test.columns if co_test[i].dtype!='O']]
    corr=df_test_num.drop(target,axis=1).corr()
    sns.heatmap(corr[(corr>=threshold)|(corr<=-(threshold))],cmap='viridis',vmax=1.0,linewidths=.1,annot=True,annot_kws={'size':10},square=True);
    plt.title('co-relation plot using testing dataset')
    plt.show()

#dropping corelated features in X_train,X_test
def corr_drop(threshold):
    global co_train,co_test,df_train9,df_test9
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = co_train.drop(target,axis=1).corr()
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    co_train.drop(col_corr,axis=1,inplace=True)
    co_test.drop(col_corr,axis=1,inplace=True)
    return col_corr
```
